```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://rynoxgear.com/collections/jackets/products/stealth-air-pro-jacket-navy-blue-light-grey-black?variant=40506607992899"
```

```python
r = requests.get(url)
htmlcontent = r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
#print(soup)
soup.prettify #Prettify the codes of the HTMl page
```

```python
title = soup.title
print(type(soup)) # BeautifulSoup
print(type(title)) # Tag
print(type(title.string)) # Navigablestring
```

```python
title=soup.title
title
```

```python
Price = soup.find_all("div",class_="selection-wrapper price product-single__price-product-template")
Price
```

```python
Product_Price = []
for i in range(0,len(Price)):
    Product_Price.append(Price[i].get_text())
Product_Price
```

```python
feature = soup.find("div",class_="product_features").text
feature.strip().replace('\n\n\n\n\n','   ')
```

```python
reviews = soup.find_all(class_="jdgm-rev-widg")
```

```python
reviews_list = []
for i in range(0,len(reviews)):
    reviews_list.append(reviews[i].get_text())
print(reviews_list)
```

```python
reviews_num = []
for i in reviews_list:
    reviews_num.append(i)
print(reviews_num)
```

```python
import pandas as pd
df = pd.DataFrame({"Product_Name":title,"Price":Product_Price,"reviews":reviews_num})
print(df)
```
